```python
import pandas as pd
```

```python
heart = pd.read_csv("./heart_X.csv")
heart = heart.drop("Unnamed: 0", axis=1)
heart.head()
```

```python
heart["Sex"].replace(to_replace=1, value="male", inplace=True)
heart["Sex"].replace(to_replace=0, value="female", inplace=True)
```

```python
heart.head()
```

```python
heart["cp"].replace(to_replace=1, value="type_1", inplace=True)
heart["cp"].replace(to_replace=2, value="type_2", inplace=True)
heart["cp"].replace(to_replace=3, value="type_3", inplace=True)
heart["cp"].replace(to_replace=4, value="type_4", inplace=True)
```

```python
heart.head()
```

```python
heart.to_csv("./heart_X.csv")
```
